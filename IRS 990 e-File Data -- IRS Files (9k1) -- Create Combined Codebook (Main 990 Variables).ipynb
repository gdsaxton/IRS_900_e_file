```python
def agg_funcs(x):
    names = {
        'description': x['description'].head(1).values[0],
        'location_code': x['location_code'].head(1).values[0],        
        'data_type_xsd': x['data_type_xsd'].head(1).values[0],
        'python_data_type': x['python_data_type'].head(1).values[0],
        #'PARSING NOTES': x['PARSING NOTES'].head(1).values[0],       
        #'OTHER NOTES': x['OTHER NOTES'].head(1).values[0],       
    }
    #THE FOLLOWING SHORTCUT WORKS BUT CHANGES THE ORDER OF THE COLUMNS
    #return pd.Series(names, index = list(names.keys()))
    return pd.Series(names, index=['description', 'location_code', 'data_type_xsd', 'python_data_type', 
                                  #'PARSING NOTES', 'OTHER NOTES'
                                  ])

#new_variables_df = concordance.groupby(['variable_name_new']).apply(agg_funcs)

new_variables_df = (
    concordance
    .groupby(['variable_name_new'])
    .apply(agg_funcs, include_groups=False)
    .reset_index()
)

#new_variables_df = new_variables_df.reset_index()
print('# of variables:', len(new_variables_df))
new_variables_df[:]
```

#### Check if any variables from the concordance file are missing in the dataset and then delete from `new_variables_df`

```python
set(new_variables_df['variable_name_new'].tolist()) - set(df.columns.tolist())
```

```python
# Define the set of variables to remove
vars_to_remove = {
    'F9_09_EXP_OTH_TOT',
    'F9_10_NAFB_RESTRICT_PERM_BOY',
    'F9_10_NAFB_RESTRICT_TEMP_BOY',
    'F9_10_NAFB_UNRESTRICT_BOY'
}

# Keep only rows NOT in the set
new_variables_df = new_variables_df[~new_variables_df['variable_name_new'].isin(vars_to_remove)].reset_index(drop=True)
```

#### Check which variables are in the dataset but missing from concordance file

```python
set(df.columns.tolist()) - set(new_variables_df['variable_name_new'].tolist())
```

```python
# Compute the set of missing variables
missing_vars = sorted(set(df.columns.tolist()) - set(new_variables_df["variable_name_new"].tolist()))

print("Variables to add:", len(missing_vars))
```

```python
# Create a DataFrame of missing variables with placeholder descriptions
new_rows = pd.DataFrame({
    "variable_name_new": missing_vars,
    "description": ["DESCRIPTION NEEDED"] * len(missing_vars)
})

# Append to your existing DataFrame
new_variables_df = pd.concat([new_variables_df, new_rows], ignore_index=True)

# Confirm everything is there now
set(df.columns.tolist()) - set(new_variables_df["variable_name_new"].tolist())
```

#### Add known descriptions

```python
# dictionary of known descriptions
known_descriptions = {
    "501c3": "Binary indicator of 501c3 status",
    "DLN": "Unique identifier for filing",
    "EIN": "EIN",
    "F9_00_HD_TIME_STAMP_yr": "Year when return created (generated from F9_00_HD_TIME_STAMP)",
    "OrganizationName": "Organization Name",
    "URL": "Unique identifier for filing -- use for merging with filings-level datasets (also for seeing raw filing online)",
    "ein_int": "Integer version of EIN",
    "filing_year_had_duplicate": "Indicator of >=1 filing for org/year (GENERATED VIA DUPLICATE CHECK)",
    "fiscal_year": "Fiscal year"
}

# update descriptions in-place
new_variables_df["description"] = new_variables_df.apply(
    lambda row: known_descriptions.get(row["variable_name_new"], row["description"]),
    axis=1
)
```

```python
# Make a copy of the column to avoid SettingWithCopyWarning
col = new_variables_df["variable_name_new"]

# Rule 1: startswith "SJ_01_"
new_variables_df.loc[col.str.startswith("SJ_01_"), "description"] = "See Schedule J (Part I) notebook"

# Rule 2: contains "SJ_02"
new_variables_df.loc[col.str.contains("SJ_02", na=False), "description"] = "See Schedule J (Part II) notebook"

# Rule 3: startswith "BMF_"
new_variables_df.loc[col.str.startswith("BMF_"), "description"] = "see BMF notebook"

# Rule 4: BusinessName
new_variables_df.loc[col == "BusinessName", "description"] = "Name of organization"

# Rule 5: BusinessNameControlTxt
new_variables_df.loc[col == "BusinessNameControlTxt", "description"] = "Name of organization"

# Rule 6: in ['ForeignAddress', 'ForeignPhoneNum', 'InCareOfName', 'InCareOfNm']
address_vars = ['ForeignAddress', 'ForeignPhoneNum', 'InCareOfName', 'InCareOfNm', 'Name', 'NameControl',
               'Phone', 'PhoneNum']
new_variables_df.loc[col.isin(address_vars), "description"] = "Parsed 990 address"

# Rule 7: COUNTY_CODE
new_variables_df.loc[col == "COUNTY_CODE", "description"] = "County FIPS code"

# Rule 7: ZIP5
new_variables_df.loc[col == "ZIP5", "description"] = "Main ZIP code variable"

# Rule 7: USAddress
new_variables_df.loc[col == "USAddress", "description"] = "Unparsed USAddress column (dictionary column)"

# Rule 8: _id
new_variables_df.loc[col == "_id", "description"] = "ignore this variable (I think it is MongoDB identifier)"

# Rule 8: NTEE
new_variables_df.loc[col == "NTEE", "description"] = "3-digit NTEE code"

# Rule 9: NTEE_MAJ12
new_variables_df.loc[col == "NTEE_MAJ12", "description"] = "Industry Fixed Effects Variable (12-category) - old names"

# Rule 10: NTEE_MAJ12_EV
new_variables_df.loc[col == "NTEE_MAJ12_EV", "description"] = "Industry Fixed Effects Variable (12-category) - new names"
```
